```python

import numpy as np
import matplotlib.pyplot as plt
import re
from gensim.models import KeyedVectors
import warnings
warnings.filterwarnings("ignore")
```

```python
cn_model = KeyedVectors.load_word2vec_format('embeddings/zh_wiki_word2vec_300.txt', 
                                             binary=False, unicode_errors="ignore")
```

```python
embedding_dim = cn_model['吳'].shape[0]
print('词向量的长度为{}'.format(embedding_dim))
cn_model['吳']
```

```python
cn_model.most_similar(positive=['放電'], topn=10)
```

```python
cn_model.similarity('男人', '女人')
```

```python
test_words = '老師 會計師 程序員 律師 醫生 老人'
test_words_result = cn_model.doesnt_match(test_words.split())
print('在 '+test_words+' 中:\n不是同一类别的词为: %s' %test_words_result)
```
